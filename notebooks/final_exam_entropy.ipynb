```python
import math
```

```python
logits = []
logits.append([2.29, -2.54, 0.02, -0.45, 0.52, -0.69])
logits.append([-0.39, -1.18, 0.95, 0.75, -1.20, -0.88])
logits.append([-1.17, 1.00, 0.94, 0.04, -0.12, -2.40])
```

```python
targets = [0, 3, 2]
```

```python
logits
```

```python
def softmax(logits):
    p = []
    for row in logits:
        row_exp = [math.exp(log) for log in row]
        
        row_total = sum(row_exp)
    
        row_probs = [val / row_total for val in row_exp]
        
        p.append(row_probs)
    
    return p

probs = softmax(logits)
probs
```

```python
sum(probs[0]), sum( probs[1]), sum(probs[2])
```

```python
# Argmax to get predictions
model_predictions = [ row.index(max(row)) for row in probs ]
model_predictions
```

```python
# Cross entropy losses
# for each row inth batch, get the probability asigned to the correct token then compute the cross entropy loss for that example
def cross_entropy(probs, targets):
    losses = []
    for i in range(len(probs)):
        correct_token_prob = probs[i][targets[i]]
        loss = -math.log(correct_token_prob)
        losses.append(loss)
    return losses

losses = cross_entropy(probs, targets)
losses
```

```python
# Calculate average loss
ans = sum(losses) / len(losses)
ans
```
