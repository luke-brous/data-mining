```python
import math
```

```python
logits = []
logits.append([1.9, -2.1, -0.3, 0.4, -0.8])
logits.append([0.3, 2.5, -0.5, 0.0, -1.1])
logits.append([-0.6, 0.2, 1.4, -0.9, -0.2])
```

```python
logits
```

```python
def softmax(logits):
    probs = [ [math.exp(log) for log in row] for row in logits ]
    
    total = sum(val for row in probs for val in row)
    
    probs = [ [log / total for log in row] for row in probs ]
    
    return probs

ret = softmax(logits)
ret
```
