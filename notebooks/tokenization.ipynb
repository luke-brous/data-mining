# Tokenizing a story

```python
with open("gautier.txt", "r") as f:
    raw_text = f.read()
print(raw_text[:100])
```

```python
tokens = re.split(r'([.?!:"(,)]|\s)', raw_text) # keep punctuation as separate tokens
tokens = [item for item in tokens if item.split()] # remove empty tokens
tokens = sorted(list(set(tokens)))  # unique tokens only 
print(tokens[100:])
```

```python
import re

class SimpleTokenizer:

    
    def __init__(self, vocab):
        if "<|unk|>" not in vocab:
            vocab["<|unk|>"] = len(vocab)
        self.str_to_int = vocab
        self.int_to_str = {index: token for token, index in vocab.items()}

    def encode(self, text):
        tokens = re.split(r'([,.?_!:"()]|\s)', text)
        tokens = [item for item in tokens if item.strip()]
        tokens = [
           item if item in self.str_to_int else "<|unk|>" for item in tokens if item.split()
        ]
        ids = [self.str_to_int[token] for token in tokens]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
```

```python
if __name__ == "__main__":
    tokenizer = SimpleTokenizer(vocab)
    text = "The most intense silence prevailed in the apartment; I had stopped the clock so that I might not hear the ticking of the pendulum,"
    ids = tokenizer.encode(text)
    print(ids)
```

```python
decoded_text = tokenizer.decode(ids)
print(decoded_text)
```
