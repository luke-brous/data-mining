```python
import math
```

```python
x = [1.9, -2.1, -0.3, 0.4, -0.8]
```

```python
# use the exp func to raise e to a given power
math.exp(1.9)
```

```python
def softmax(v):
    p = []
    for i in range(len(v)):
        p.append(math.exp( v[i] ))
    total = sum(p)
    for i in range( len(v) ):
        p[i] = p[i] / total
        # p[i] = round(p[i], 4) # for rounding
    return p
```

```python
probs = softmax(x)
probs
```

```python
sum(probs)
```

```python
# for this set of logits, the correct next token has ID 1
# so the cross entropy loss is -log( probs[1] ) :
L = -math.log( probs[1] )
L
```
